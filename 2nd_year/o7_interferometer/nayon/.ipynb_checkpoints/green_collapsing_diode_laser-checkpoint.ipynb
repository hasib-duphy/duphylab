```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

### Green Diode Laser
### Green Laser wavelength (collapsing)
#### $$λ_{green\_collapsing} = \frac{2 \Delta D}{fn}$$
#### $$ \Delta D = \frac{fλ}{2} n $$

```python
f = 20
data = pd.DataFrame()
data['n'] = np.arange(0,241,20)
data['D (mm)'] = [24.35, 24.53, 24.73, 24.85, 24.98, 25.12,
                 25.25, 25.38, 25.53, 25.71, 25.85,
                 26.03, 26.19]
data['ΔD (mm)'] = data['D (mm)'] - data.loc[0,'D (mm)']

data.to_csv('green_collapsing_diode_laser.csv', index = False)
data
```

#### $$y = mx$$ 
#### $$regression$$

```python
x = data['n'].to_numpy()
y = data['ΔD (mm)'].to_numpy()*1e-3

def formula(n,λ):
    return 0.5*f*λ*n

para, para_cov = curve_fit(formula, x, y, [500e-9])
λ = para[0]
print('λ =', λ)

x_reg = np.linspace(0,250,125)
y_reg = formula(x_reg, λ)

plt.plot(x,y,'.', color = "red", label = "Data")
plt.plot(x_reg,y_reg, '-', color = "green", label = "Regressed")
plt.legend()
plt.show()
```
